```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/Projects/D9/grades_withnulls.csv')
```

```python
x=data[['SEM1','SEM2','SEM3']]
y=data['Grade']
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
k=3
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(x,y)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
data_for_model = df[['SEM3', 'Grade']].dropna()
X = data_for_model[['SEM3']]
y = data_for_model[['Grade']]
LR = LinearRegression()
LR.fit(X, y)
plt.scatter(X, y, alpha=0.5)
x_line = np.linspace(X.min(), X.max(), 100)
y_line = LR.predict(x_line)
plt.plot(x_line, y_line, linewidth=3, color='red')
plt.xlabel("SEM3")
plt.ylabel("Grade")
plt.title("Linear Regression: SEM3 vs Grade")
plt.show()
```

```python
n_d=np.array([[9.0, 9.5, 9.2]]) # Example: Added a third feature and adjusted values to be in a sensible range.
prediction=knn.predict(n_d)
if prediction >= 9.5:
  print("Grade A+")
elif prediction >= 9:
  print("Grade A")
elif prediction >= 8.5:
  print("Grade B+")
elif prediction >= 8:
  print("Grade B")
elif prediction >= 7.5:
  print("Grade C")
else:
  print("Poor score")
```
